```python
import pandas as pd
from pymcprophet.model import BayesTS, BayesTSConfig
from prophet import Prophet
```

```python
df = df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/main/examples/example_wp_log_peyton_manning.csv')
df['ds'] = pd.to_datetime(df['ds'])
```

```python
len(set(df["y"].unique()))
```

## Model Matrix

```python
# ts_config = BayesTSConfig(daily_seasonality="disabled", weekly_seasonality="disabled", yearly_seasonality="disabled")
ts_config = BayesTSConfig()
ts_model = BayesTS(ts_config)
ts_model.assign_model_matrix(df)

ts_model.model_df
```

```python
ts_model.model_df
```

## Logistic

```python
df_logistic = df.copy()
df_logistic["cap"] = 10

ts_config = BayesTSConfig(growth="logistic", seasonality_mode="multiplicative")
ts_model = BayesTS(ts_config)
ts_model.assign_model_matrix(df_logistic)

ts_model.model_df
```

## Holidays

```python
playoffs = pd.DataFrame({
  'holiday': 'playoff',
  'ds': pd.to_datetime(['2008-01-13', '2009-01-03', '2010-01-16',
                        '2010-01-24', '2010-02-07', '2011-01-08',
                        '2013-01-12', '2014-01-12', '2014-01-19',
                        '2014-02-02', '2015-01-11', '2016-01-17',
                        '2016-01-24', '2016-02-07']),
  'lower_window': 0,
  'upper_window': 1,
})
superbowls = pd.DataFrame({
  'holiday': 'superbowl',
  'ds': pd.to_datetime(['2010-02-07', '2014-02-02', '2016-02-07']),
  'lower_window': 0,
  'upper_window': 1,
})
holidays = pd.concat((playoffs, superbowls))
holidays
```

```python
train_df = df.query("ds.dt.year < 2014")
test_df = df.query("ds.dt.year > 2014")



ts_config = BayesTSConfig()
ts_model = BayesTS(ts_config)
ts_model.add_holiday_country("US")
ts_model.add_holiday(holidays)
ts_model.assign_model_matrix(train_df)

ts_model.model_df
```

```python
ts_model.produce_model_matrix(test_df)
```

## Seasonality

Compare Seasonality To Prophet

```python
ts_config = BayesTSConfig()
ts_model = BayesTS(ts_config)
ts_model.assign_model_matrix(df)

ts_model_seasonality_terms = ts_model.produce_model_matrix(df)[ts_model.model_spec.get_seasonality_feature_cols()]
ts_model_sine_cols = [c for c in ts_model_seasonality_terms.columns if "sin" in c]
ts_model_sine_terms = ts_model_seasonality_terms[ts_model_sine_cols]

propet_model =  Prophet()
transformed = propet_model.preprocess(df)
prophet_seasonality_terms = transformed.X

prophet_sine_cols = [c for c in prophet_seasonality_terms.columns if int(c.split("_")[-1]) % 2]
prophet_sine_terms = prophet_seasonality_terms[prophet_sine_cols]

max_dif = (ts_model_sine_terms.values - ts_model_sine_terms.values).max()

print(max_dif)
```
